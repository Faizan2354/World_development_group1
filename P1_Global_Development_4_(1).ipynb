```python
#Handling Outliers(Capping)
for col in df.select_dtypes(include=['int64','float64']).columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    df[col] = np.where(df[col] > upper, upper, df[col])
    df[col] = np.where(df[col] < lower, lower, df[col])
```

```python
# ============================================================
# MODEL BUILDING AND EVALUATION
# World Development Dataset - Clustering Analysis
# ============================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.decomposition import PCA
from sklearn.metrics import (silhouette_score, davies_bouldin_score,
                             calinski_harabasz_score)
from sklearn.cluster import (KMeans, AgglomerativeClustering,
                             DBSCAN, MeanShift, estimate_bandwidth)
from sklearn.mixture import GaussianMixture
from scipy.cluster.hierarchy import dendrogram, linkage
from scipy.spatial.distance import cdist
```

```python
# ─────────────────────────────────────────────
# STEP 1 ▸ REPRODUCE PREPROCESSING
# ─────────────────────────────────────────────
df = pd.read_excel("C:/Users/Faizan/Downloads/P659_World_development_dataset.xlsx")

cols_to_clean = ['GDP', 'Tourism Inbound', 'Tourism Outbound',
                 'Health Exp/Capita', 'Business Tax Rate']
for col in cols_to_clean:
    df[col] = df[col].replace('[\$,%]', '', regex=True)
    df[col] = pd.to_numeric(df[col], errors='coerce')

df.drop(columns=['Ease of Business'], inplace=True, errors='ignore')
df.fillna(df.median(numeric_only=True), inplace=True)
```

```python
# Outlier capping (IQR method)
for col in df.select_dtypes(include=['int64', 'float64']).columns:
    Q1, Q3 = df[col].quantile(0.25), df[col].quantile(0.75)
    IQR = Q3 - Q1
    df[col] = np.clip(df[col], Q1 - 1.5 * IQR, Q3 + 1.5 * IQR)
```

```python
# New feature
df['Avg Life Expectancy'] = (df['Life Expectancy Female'] +
                              df['Life Expectancy Male']) / 2

print("Dataset shape after preprocessing:", df.shape)
```

```python
# ─────────────────────────────────────────────
# STEP 2 ▸ FEATURE SELECTION & SCALING
# ─────────────────────────────────────────────
# Select meaningful numerical features for clustering
feature_cols = [
    'GDP', 'Health Exp/Capita', 'Internet Usage',
    'Birth Rate', 'Life Expectancy Female', 'Life Expectancy Male',
    'Avg Life Expectancy', 'CO2 Emissions', 'Business Tax Rate',
    'Tourism Inbound', 'Tourism Outbound'
]

# Keep only columns that exist in the dataframe
feature_cols = [c for c in feature_cols if c in df.columns]
print(f"\nFeatures used for clustering ({len(feature_cols)}):", feature_cols)

X = df[feature_cols].copy()
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# PCA for 2D visualisation
pca = PCA(n_components=2, random_state=42)
X_pca = pca.fit_transform(X_scaled)
print(f"\nPCA explained variance ratio: {pca.explained_variance_ratio_}")
print(f"Total variance captured: {pca.explained_variance_ratio_.sum():.2%}")
```

```python
# ─────────────────────────────────────────────
# STEP 3 ▸ OPTIMAL K — ELBOW + SILHOUETTE
# ─────────────────────────────────────────────
print("\n" + "="*60)
print("FINDING OPTIMAL NUMBER OF CLUSTERS")
print("="*60)

k_range = range(2, 11)
inertia_values, silhouette_values = [], []

for k in k_range:
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = km.fit_predict(X_scaled)
    inertia_values.append(km.inertia_)
    silhouette_values.append(silhouette_score(X_scaled, labels))

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle("Optimal K Selection", fontsize=14, fontweight='bold')

axes[0].plot(k_range, inertia_values, 'bo-', linewidth=2, markersize=6)
axes[0].set_xlabel("Number of Clusters (k)")
axes[0].set_ylabel("Inertia (WCSS)")
axes[0].set_title("Elbow Method")
axes[0].grid(alpha=0.3)

axes[1].plot(k_range, silhouette_values, 'ro-', linewidth=2, markersize=6)
axes[1].set_xlabel("Number of Clusters (k)")
axes[1].set_ylabel("Silhouette Score")
axes[1].set_title("Silhouette Score vs k")
axes[1].grid(alpha=0.3)

plt.tight_layout()
plt.savefig("optimal_k_selection.png", dpi=150, bbox_inches='tight')
plt.show()

best_k = silhouette_values.index(max(silhouette_values)) + 2
print(f"\nOptimal k (highest silhouette): {best_k}")
```

```python
# ─────────────────────────────────────────────
# STEP 4 ▸ MODEL 1 — K-MEANS CLUSTERING
# ─────────────────────────────────────────────
print("\n" + "="*60)
print("MODEL 1: K-MEANS CLUSTERING")
print("="*60)

kmeans = KMeans(n_clusters=best_k, random_state=42, n_init=10)
df['KMeans_Cluster'] = kmeans.fit_predict(X_scaled)

sil_km  = silhouette_score(X_scaled, df['KMeans_Cluster'])
db_km   = davies_bouldin_score(X_scaled, df['KMeans_Cluster'])
ch_km   = calinski_harabasz_score(X_scaled, df['KMeans_Cluster'])

print(f"  Clusters Formed    : {best_k}")
print(f"  Silhouette Score   : {sil_km:.4f}  (higher = better, max 1)")
print(f"  Davies-Bouldin     : {db_km:.4f}  (lower  = better)")
print(f"  Calinski-Harabasz  : {ch_km:.2f}  (higher = better)")
print(f"  Inertia (WCSS)     : {kmeans.inertia_:.2f}")

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle("Model 1: K-Means Clustering", fontsize=13, fontweight='bold')

scatter = axes[0].scatter(X_pca[:, 0], X_pca[:, 1],
                          c=df['KMeans_Cluster'], cmap='tab10', alpha=0.7, s=50)
axes[0].set_title(f"PCA Plot (k={best_k})")
axes[0].set_xlabel("PC1"); axes[0].set_ylabel("PC2")
plt.colorbar(scatter, ax=axes[0], label='Cluster')

cluster_counts = df['KMeans_Cluster'].value_counts().sort_index()
axes[1].bar(cluster_counts.index, cluster_counts.values,
            color=plt.cm.tab10(np.linspace(0, 1, best_k)))
axes[1].set_title("Countries per Cluster")
axes[1].set_xlabel("Cluster"); axes[1].set_ylabel("Count")

plt.tight_layout()
plt.savefig("model1_kmeans.png", dpi=150, bbox_inches='tight')
plt.show()
```
